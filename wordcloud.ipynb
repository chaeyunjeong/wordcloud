```python
from collections import Counter
from konlpy.tag import Okt

import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
```

```python
fff = pd.read_csv("https://raw.githubusercontent.com/chaeyunjeong/wordcloud/master/fff.csv")
ggg = pd.read_csv("https://raw.githubusercontent.com/chaeyunjeong/wordcloud/master/ggg.csv")
hhh = pd.read_csv("https://raw.githubusercontent.com/chaeyunjeong/wordcloud/master/hhh.csv")
iii = pd.read_csv("https://raw.githubusercontent.com/chaeyunjeong/wordcloud/master/iii.csv")
jjj = pd.read_csv("https://raw.githubusercontent.com/chaeyunjeong/wordcloud/master/jjj.csv")

ignore = ['청년', '고용', '정책', '취업', '일자리', '실업', '고용노동부', '연구',
         '관', '및', '층', '위', '목적', '것', '수', '이', '그', '위해', '대한',
          '등', '명', '개', '세', '률', '중', '통해', '이번', '제', '고', '를',
         '한국', '정부', '대책', '우리나라', '회', '설명', '증가', '인', '이상',
          '또한']
```

```python
ffftext = ' '.join(fff.iloc[:,0])
nlp = Okt()
fffnouns = nlp.nouns(ffftext)
fffcount = Counter(fffnouns)

for word in list(fffcount):
  if word in ignore:
    del fffcount[word]

print(fffcount.most_common(50))
```

```python
jjjtext = ' '.join(jjj.iloc[:,0])
nlp = Okt()
jjjnouns = nlp.nouns(jjjtext)
jjjcount = Counter(jjjnouns)

for word in list(jjjcount):
  if word in ignore:
    del jjjcount[word]

print(jjjcount.most_common(50))
```

```python
#fff으로 만든 wordcloud
wordcloudfff = WordCloud(font_path = '/usr/share/fonts/truetype/nanum/NanumGothicEco.ttf', background_color="white", max_words=20)
wordcloudfff.generate_from_frequencies(frequencies=fffcount)

plt.imshow(wordcloudfff, interpolation='bilinear')
plt.axis("off")
plt.show()
```
